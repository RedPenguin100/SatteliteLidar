```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_hdf('./data/data.h5')
```

```python
bad_data = data[data['BEAM'] == 3]

def plot_scatter(data: pd.DataFrame, beam):
    plt.scatter(data['rh_98'], data['als_h'])
    plt.xlabel('rh_98')
    plt.ylabel('als_h')
    plt.title(f'BEAM{beam}')
    plt.show()

plot_scatter(bad_data, 3)
```

```python
for beam in range(0, 8):
    plot_scatter(data[data['BEAM'] == beam], beam)
```

```python
rh_98 = data['rh_98'].to_numpy()
rh_99 = data['rh_99'].to_numpy()
rh_100 = data['rh_100'].to_numpy()
x = np.arange(len(rh_98))
print(len(rh_100))

stacked = np.vstack([rh_98, rh_99, rh_100]) / rh_100
variance_vector = np.var(stacked, axis=0) * 100
np.set_printoptions(precision=3)

def sigmoid(x):
    return 1 / (1  + np.exp(-x))

def custom(x):
    return np.arctan(x) * 2 / np.pi

data['variance'] = variance_vector
data['variance_sigmoid'] = sigmoid(variance_vector)
data['variance_custom'] = custom(variance_vector)


sorted_data = data.sort_values(by='variance')
sorted_data[['variance', 'variance_sigmoid', 'variance_custom', 'rh_98', 'rh_99', 'rh_100', 'sensitivity']]
```
